# What are operators ?

Operators are mathematical symbols that are used to perform various operations between operands. The numbers on which the operation is performed are called **operands** and the symbol that performs the operation is called an **operator**

# Arithmetic operators

These operators are used to perform mathematical operations

```python
num1 = 32
num2 = 24
res = num1 + num2
print("The sum of the numbers is",res)
```

```python
num1 = 32
num2 = 24
res = num1 - num2
print("The difference between the numbers is",res)
```

```python
num1 = 32
num2 = 24
res = num1 * num2
print("The product of the numbers is",res)
```

```python
num1 = 32
num2 = 4
res = num1 / num2
print("The quotient after dividing",num1,"by",num2, "is",res)
print("The type of",res,"is",type(res))
```

```python
num1 = 32
num2 = 8
res = num1 // num2
print("The quotient after dividing",num1,"by",num2, "is",res)
```

```python
num1 = 132
num2 = 48
res = num1 % num2
print("The remainder after dividing",num1,"by",num2, "is",res)
```

```python
num = 32
res = num ** 5
print(num,"raised to the power of 5 is",res)
```

# Assignment operators

Assignment operators are used to assign values to variables. They take the right value and assign it to the left side variable.

```python
a = 10
print(a)
```
