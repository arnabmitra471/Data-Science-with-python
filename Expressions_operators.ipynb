# What are operators ?

Operators are mathematical symbols that are used to perform various mathematical and logical operations between operands. The numbers on which the operation is performed are called **operands** and the symbol that performs the operation is called an **operator**

# Arithmetic operators

These operators are used to perform mathematical operations

```python
num1 = 32
num2 = 24
res = num1 + num2
print("The sum of the numbers is",res)
```

```python
num1 = 32
num2 = 24
res = num1 - num2
print("The difference between the numbers is",res)
```

```python
num1 = 32
num2 = 24
res = num1 * num2
print("The product of the numbers is",res)
```

```python
num1 = 32
num2 = 4
res = num1 / num2
print("The quotient after dividing",num1,"by",num2, "is",res)
print("The type of",res,"is",type(res))
```

```python
num1 = 32
num2 = 8
res = num1 // num2
print("The quotient after dividing",num1,"by",num2, "is",res)
```

```python
num1 = 132
num2 = 48
res = num1 % num2
print("The remainder after dividing",num1,"by",num2, "is",res)
```

```python
num = 32
res = num ** 5
print(num,"raised to the power of 5 is",res)
```

# Relational Operators

These operators are used to compare two values to check for a condition. These operators return `True` or `False` based on whether the condition gets satisfied or not. These operators are used to check various conditions for `if else` statements.

```python
num1 = 12
num2 = 12
print(num1 > num2)
print(num1 < num2)
print(num1 >= num2)
print(num1 <= num2)
print(num1 == num2)
print(num1 != num2)
```

# Logical Operators

There are three logical operators in python which are as follows. Logical operators are used to form compound conditions. These conditions can be used in conditional statements.

1. Logical AND `and`
2. Logical OR `or`
3. Logical NOT `not`

## Logical `and` operator

The logical `and` operator checks two or more conditions and returns `True` if and only if all the conditions are `True`. If one condition is `False` then the entire condition evaluates to `False`

```python
num1 = False
num2 = False
print(num1 and num2)
```

## Logical `or` operator

The logical `or` operator checks two or more conditions and returns `True` if any one condition is `True`. If all the conditions are `False` then only the entire condition evaluates to `False`

```python
cond1 = False
cond2 = False
print(cond1 or cond2)
```

# Logical `not` operator

The logical not operator inverts the logic of a particular condtion. If any condition evaluates to `True` then the logical `not` operator will return `False` and vice versa.

```python
cond = False
print(not cond)
```

# Bitwise operators

These operators first convert the integers to their corresponding binary representations and then apply these operations. The result is then converted back to an integer.

Python supports the following bitwise operators

* Bitwise AND `&`
* Bitwise OR `|`
* Bitwise NOT `~`
* Bitwise XOR `^`
* Bitwise left shift `<<`
* Bitwise right shift `>>`

```python
c = 34
d = 32
print(c & d)
```

```python
c = 34
d = 32
print(c | d)
```

```python
c = 34
print(~c)
```

```python
c = 34
d = 32
print(c ^ d)
```

```python
num1 = 21
lshift_res = num1 << 2
print(lshift_res)
```

```python
num1 = 21
rshift_res = num1 >> 2
print(rshift_res)
```

# Membership operators

These operators are used to check whether an item is present in the sequence or not. The membership operators in python are as follows
* `in` operator
* `not in` operator

```python
l1 = [23,56,43,24,22]

print(23 in l1) # Returns True as 23 is in the list
print(56 in l1) # Returns True as 56 is in the list
print(123 in l1) # in operator returns False as 123 is not in the list
print(123 not in l1) # not in operator returns False as 123 is not in the list
```
