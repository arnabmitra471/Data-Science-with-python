## Comments in python

Comments are a piece of text that the python interpreter will ignore. It won't execute the lines which are comments

```python
# Hey !! I am comment and I will not be executed by the python interpreter
print("I am back !!")
# print("How is it going !!")
print("I love python a lot !!")
```

```python
"""
Hey !! This is a
multiline comment
"""
print("These are some more lines of code")
```

```python
def my_func():
    """This function prints
    a greeting message for the users"""
    print("Hello from the function !!")

my_func()
```
