# Lists and tuples

## Tuples

* Tuples are an ordered sequence of elements
* Tuples are immutable in python which means they can't be changed in place once they are created

````python
ratings = (10,9,10,8,5,6,9,5,6,6,10,4,3,6,7,8,10,9,5)
print(ratings)
print(type(ratings))
print(len(ratings))
````

````python
print(ratings[0])
print(ratings[1])
print(ratings[2])
print(ratings[3])
````

# Concatenating two tuples using the `+` operator

Python actully does not add those two tuples together. Instead it creates a brand new tuple in the memory and copies the references of the elements from the tuples `fr1` and `fr2`

````python
fr1 = ("Sristi","Rajeswari","Aneesh","Mamun")
fr2 = ("Alice","Bob","Evan","Dan")
all_friends = fr1 + fr2
print(all_friends)
print(id(fr1))
print(id(all_friends))
````

````python
t2 = (56,76,89,23,98,12,34,76,100,178,156,143)
print(t2)
# print(type(t2))
````

# Tuple slicing in python

If we want to access a particular subset of elements from a tuple we can use tuple slicing. The syntax for tuple slicing is as follows
`tuple_name[start:end:step]`

```python
t2 = (56,76,89,23,98,12,34,76,100,178,156,143)
print(t2[2:7])
```

````python
t2 = (56,76,89,23,98,12,34,76,100,178,156,143)
print(t2[2:7])
print(t2[-6:-2])
print(t2[6:10])
````

# Sorting a tuple

We can sort tuples,lists,strings,etc by using the function `sorted()`

````python
sorted_ratings = sorted(ratings,reverse=True)
print(sorted_ratings)
````

````python
words = ["Apple","Kiwi","Banana","Marshmallow","Watermelon","Lemon","Mango"]
sorted_fruits = sorted(words,key=len)
print(sorted_fruits)
````

# Tuple of tuples (Nested Structures)

Tuples in python can also contain other tuples consisting of elements of different types

````python
n_tup = (1,2,("Pop","Rock"),(3,4),("Disco",(1,2)))
print(n_tup)

for element in n_tup:
    if isinstance(element,(tuple,str)):
        for sub_element in element:
            print(sub_element)
    else:
        print(element)
````

````python
tup3 = (1,3,4,5,89,14,56,"Mango","Banana","Apple","Pear")
new_tup = tup3
print(tup3)
print(new_tup)

print(id(tup3))
print(id(new_tup))

print(tup3[0])
# tup3[0] = 34
````

## Lists

* Lists are an ordered collection of items
* They are created by writing comma separated values enclosed in square brackets `[]`
* They are mutable in nature. This means that once a list is created and allocated memory, we can make changes at the same memory address to the original list that was created.
* The only major difference between lists and tuples is the concept of mutability

The `list()` function - This is a built in typecasting method which allows us to convert any variable from any other data type into the `list` data type

````python
l1 = list(t2)
print(l1)
print(type(l1))
````

# List indexing

We can access the elements of a list by indexing the list. The index defines the specific position of an element in a list
Indexing in python starts from 0 and goes upto `n-1` for a list of length n

````python
print(l1[0])
print(l1[1])
print(l1[2])
print(l1[3])
````

# Iterating over a list using the `for` loop

We can iterate over list using the `for` loop to access its elements as follows

````python
for i in l1:
    print(i)
````

We can also print the elements of the list in reverse order using the for loop

# Detemining the length of a list
We can determine the length of a list using the built in `len()` function as follows. The length of a list indicates the number of elements in the list

````python
print(len(l1))
````

# List slicing

If we want to access a specific portion or a subset of elements from a list, we can use list slicing in python. Slicing a list does not modify the list.

The syntax of list slicing is as follows

`list_name[start:end:step]`

Slicing a list gives us all the elements starting from the specified starting index and ranges upto but excluding the specified ending index. This means that the starting index is included but we go upto ending index - 1. The step size is optional. It is basically the jump index which specifies, how many elements to skip before taking the next element. By default the step size or jump index is 1, means picking every consecutive element before the ending index

````python
l2 = [45,78,79,45,67,18,89,78,67,54,56,123,343,216]
print(l2)
print(f"The length of the list is {len(l2)}")

# Accessing the elements starting from the first index to the 7th index
print(l2[1:8])
print(l2[1:6])
````

**Important note**

* We can also **omit** the starting index. By default slicing starts from the beginning of the list
* We can also omit the ending index. In this case slicing continues till the end of the list

````python
# Taking all the elements starting from the second index
print(l2[2:])
"""
The starting index is 0 by default. So if the starting index is omitted, slicing will start from the beginning of the list
"""
print(l2[:7])
print(l2[:7:2])
print(l2[:7:-2])
print(l2[:7:-1])
print(l2[::-1])
````
