# Strings in python

Strings in python are a sequence of characters enclosed within single or double quotes. Each character in a string has an associated index number.

```python
name = "Arnab Mitra"
print(name[0])
print(name[1])
print(name[2])
```

````python
name = "Michael Jackson"
print(name)
print(type(name))
````

# Accessing the characters of a string

We can acces the individual characters in a string by indexing because a string is like an array of characters

````python
print(name[0])
print(name[1])
print(name[2])
print(name[3])
print(name[4])
print(name[5])
````

````python
for char in name:
    print(char)
````

````python
print("The length of the sting is",len(name))
````

## Negative indexing

````python
print(name[-1])
print(name[-2])
print(name[-3])
print(name[-4])
````

# String slicing

````python
apple = "An apple a day keeps the doctor away"
print(apple[1:6])

print(name[0:4])
print(name[:4])
print(apple[1:])
print(apple[::2])
print(name[::2])
print(name[0:5:2])
````

````python
print(name + " is the best")
````

````python
first_name = "Arnab"
last_name = "Mitra"
print(first_name + " " + last_name)

print("Harry" + str(123))
````

````python
name = "Vrinda"
print(3 * name)
````

## Concept of immutability in strings

* Strings are **immutable** meaning they cannot be changed in place once they are created. We can create a new copy of the original string and then apply the modifications to that string but the original string remains unchanged.

* All the string methods return new strings thus returning a modified copy of the original string

````python
s1 = "Hello World"
print(s1[0])
upper_s1 = s1.upper()
print(upper_s1)
print(s1)
print(id(s1))
print(id(upper_s1))
````

# String methods in python

````python
studio = "Thriller is the sixth studio album"
upper_studio = studio.upper()
print(upper_studio)
````

````python
lower_studio = studio.lower()
print(lower_studio)
````

````python
print(studio.endswith("album"))
print(studio.endswith("ggg"))
````

````python
print(apple.index("day"))
print(apple.find("doctor"))
print(apple[5:20])
````

````python
print(studio.startswith("Thriller"))
print(studio.endswith("album"))
print(studio.startswith("sixth",16,26))
print(studio[16:26])
````

````python
boy = "Arnab is a good boy"
print("good" in boy)
print("123" not in boy)
````

````python
fruit = "apples-1 apples-2 apples-3"
print(fruit.count("apples"))
````

````python
apple_chars = apple.split(" ")
print(apple_chars)
````

````python
friend = "     Rohan        "
rstripped_res = friend.rstrip(" ")
lstripped_res = friend.lstrip(" ")
# Length of the original string
print(len(friend))
# Length of the string after removing trailling whitespace characters
print(len(rstripped_res))

# Length of the string after removing leading whitespace characters
print(len(lstripped_res))
````

````python
print(apple.isupper())
print(upper_studio.isupper())
````

````python
print(apple.islower())
print(lower_studio.islower())
````

````python
num_alpha = "213yrgfg67221"
print(num_alpha.isalnum())
````

````python
pure_alpha = "abhdhgcghdcghc"
print(pure_alpha.isalpha())
````

````python
digit_string = "12373737373jhdfhudfgdfAAShdhHH"
print(digit_string.isdigit())
````

````python
blog_heading = "intRoducTion To JS"
cap_str = blog_heading.capitalize()
print(cap_str)
````

````python
fav_lang = "Python Is My Favorite Language"
case_switched_str = fav_lang.swapcase()
print(case_switched_str)
````

````python
space_str = "     "
print(space_str.isspace())
````
