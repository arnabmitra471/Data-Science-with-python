# Type casting
The process of conversion of a variable from one datatype to another is known as **typecasting** in python. It is also known as **type conversion**

## Forms of typecasting
1. Implicit typecasting
2. Explicit typecasting

### Implicit typecasting

```python
num3 = 34.5
num4 = 43 # Python implicitly converts 43 to a float 43.0 and then adds num3 to it 
print(f"The sum of {num3} and {num4} is {num3 + num4}")
```

### Explicit typecasting

```python
num2 = "12"
num3 = "45"
int_num2 = int(num2)
int_num3 = int(num3)
print(type(num2))
print(type(num3))
print(int_num2 + int_num3)
print(type(int_num2),type(int_num3))
```

## `complex` data type

```python
comp1 = 36 - 45j
comp2 = 26 - 15j
comp_sum = comp1 + comp2
comp_prod = comp1 * comp2
print("The sum of the numbers is",comp_sum)
print("The product of the numbers is",comp_prod)
print("The type of",comp1,"is",type(comp1))
print("The type of",comp2,"is",type(comp2))
print(comp1.real)
print(comp1.imag)
```

# Concept of mutability and immutability

```python
items_tuple = (1,2,3,4,5)
print(items_tuple)
print("The id of the first items tuple is",id(items_tuple))
items_list = list(items_tuple)
items_list.append(23)
updated_tuple = tuple(items_list)
print(updated_tuple)
print("The id of the updated items tuple is",id(updated_tuple))
```

```python
items_list = [1,2,3,4,5]
print(items_list)
print("The id of the initial items list is",id(items_list))
items_list.append(23)
print(items_list)
print("The id of the updated items list is",id(items_list))
```

# `list` datatype

```python
l1 = [65,78,86,89,34,123,764]
print(l1)
print(type(l1))
print(l1[0])
print(l1[1])
l1[0] = 543
l1[1] = 321
print(l1[1:5])
```

# `tuple` datatype

```python
t1 = (65,86,23,78,1233,77,233)
print(t1)
print(t1[0])
print(t1[1])
print(t1[2])
print(t1[3])
# t1[1] = 675
# Slicing a tuple
print(t1[1:4])
```

```python
s1 = "Hello World"
l2 = [32,79,45,78,65,73,132,143,221,230]
bytes_s1 = bytes(s1,encoding="utf-8")
l2_bytes = bytes(l2)
print(bytes_s1,type(bytes_s1))
print(l2_bytes)
print(type(l2_bytes))
decoded_string = bytes_s1.decode("utf-8")
print(decoded_string,type(decoded_string))
```

## Looping through a bytes object

```python
for byte in l2_bytes:
    print(byte)
```

```python
for byte in bytes_s1:
    print(byte)
```

# `bytearray` data type

```python
l3 = [43,67,78,34,12,87,89,127,222,112]
list_byte_arr = bytearray(l3)
print(list_byte_arr)

for byte in list_byte_arr:
    print(byte)

list_byte_arr[1] = 127
print(list_byte_arr)
```

# `set` data type

```python
l1 = [32,43,78,12,33]
s1 = set(l1)
print(s1,type(s1))
s1.add(132)
s1.add(453)
print(s1)
s1.clear()

s2 = set()
print(s2)
print(type(s2))
```

# `frozenset` data type

```python
l1 = [32,76,65,24,88,23]
fs1 = frozenset(l1)
print(fs1)
```

# `dict` data type

```python
d1 = {"name" : "Arnab",
      "age" : 20,
      "fav_lang" : ["Python","Javascript","C","Kotlin","Java"],
      "fav_sport" : "Chess"
     }
print(d1)
print(type(d1))
d1["fav_lang"] = ["Javascript","Python","C++"]
print(d1)
d1["can_vote"] = True
print(d1)
```
