# List methods in python

````python
l1 = [23,66,29,98,21,34,45,87,89,59,128]
print(l1)
print(type(l1))
````

# `len()` function

The `len(iterable)` function can be used to find out the length of a list

````python
n = len(l1)
print("The list l1 contains",n,"elements")
print(l1)
````

# `append()` method
Appending an element to a list means adding an element at the end of it. So the `list.append()` method is used to add an element to the end of the list

````python
l1.append(187)
print(l1)
new_length = len(l1)
print("The list l1 now contains",new_length,"elements")
````

# `copy()` method

The `list.copy()` method creates a shallow copy of a list. This is used for preventing accidental modifications to the original list.

````python
l2 = [45,76,78,90,23,12,34,90,187,123,124,[33,43,89,12,34,78,190,124,134,891,467]]
l3 = l2.copy()
l3.append(452)
print(l2)
print(l3)
````

````python
print(l3[11])

l3[11].append(490)
print(l3[11])
print(l2)
````

# Counting the occurence of a specific element in a list

````python
my_list = [2,8,7,6,7,2,3,4,5,5,9,7,7,7,12,45,24,34,35,45,45,45,90]
print(my_list)

n_occ = my_list.count(7)
print(n_occ)
````

````python
my_list.extend([47,32,89,78,81,56,23])
print(my_list)
````

# `list.index()` method

The `list.index()` method returns the index of the first occurence of an element in a list

````python
l5 = [23,83,76,65,90,123,134,176,154]
element = int(input("Enter the element you want the index of"))
print(l5.index(element))
````

## `list.pop()` method

The `list.pop()` method is used to remove the last element of the list by default or the element at a specified index in the list.

The `pop()` method returns the element that it has removed.

````python
print(l5)
element_removed = l5.pop(2)
print(element_removed)
print(l5)
````

# `list.remove()` method

* The `list.remove()` method is used to remove a specific element from a list.
* It always removes the first occurence of the element from the list
* The `list.remove()` method does not return anything, i.e it returns `None`
* The remove method raises a ValueError if the specified value is not present in the list

````python
fruits = ["Apple","Mango","Banana","Pear","Pineapple","Watermelon","Mango"]
fruits.remove("Mango")
print(fruits)
````

# `list.clear()` method

The `list.clear()` method is used to remove all the elements from the list at once and clear it all off

````python
fruits.clear()
print(fruits)
````
